### Read sheet 1

```python
# Define cols to fetch
cols = ["Fake Applicant ID", "Attendance", "Course Id"]
# Get completion data
df_sheet1 = pd.read_csv("https://s3-eu-west-1.amazonaws.com/ai-hack-q3-orion/datasets/completion_data.csv", 
                 skip_blank_lines=True,
                 index_col=1,
#                  na_values='No Data',
                 usecols=cols,
                 dtype = {"Course Id" : "object", "Fake Applicant ID" : "int32", "Attendance" : "str"}
                )
df_sheet1_filt = df_sheet1.dropna(axis=0)
df_sheet1_filt.head()
```

```python
series_attendence_total = df_sheet1_filt.groupby('Fake Applicant ID')['Attendance'].count()
series_attendence_total.head()
```

```python
series_attended = df_sheet1_filt.groupby(['Fake Applicant ID'])['Attendance'].apply(lambda x: x[x.str.contains('Attended')].count())
series_attended.head()
```

```python
df_sheet1_filt['Attendence_Perc'] = series_attended.loc[0:1] / series_attendence_total[0:1]
```

```python
df_joined['Attended_Total'] = df_sheet1_filt.groupby(['Fake Applicant ID'])['Attendance'].apply(lambda x: x[x.str.contains('Attended')].count())
```
