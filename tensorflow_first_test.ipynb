```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow import keras
```

```python
# Get app data
df_clean = pd.read_csv("https://s3-eu-west-1.amazonaws.com/ai-hack-q3-orion/datasets/final_clean_orion.csv")
```

```python
df_clean.head()
```

pd.get_dummies(df_clean,prefix=['Course ID', 'Gender', 'Income', 'Education', Primary Interest In Course', 'Hours Coded', 'How Many Hours A Week Can You Commit To Class'])

```python
train, test = train_test_split(df_clean, test_size=0.2)
trainX = train.loc[:, train.columns != 'Completed?']
trainX = trainX.sort_index()
trainY = train.loc[:, train.columns == 'Completed?']
trainY = trainY.sort_index()

testX = test.loc[:, test.columns != 'Completed?']
testX = testX.sort_index()
testY = test.loc[:, test.columns == 'Completed?']
testY = testY.sort_index()
```

```python
model = keras.Sequential([
    keras.layers.Dense(128,input_shape=(trainX.shape[1],), activation=tf.nn.relu),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(64, activation=tf.nn.relu),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(1)
])
```

```python
model.compile(optimizer='adam', 
              loss='mean_squared_error',
              metrics=['accuracy'])
```

```python
model.fit(trainX, trainY, epochs=200)
```

model.evaluate(testX, testY)
